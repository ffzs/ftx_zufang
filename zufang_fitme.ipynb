```python
import pymongo
import pandas as pd
import numpy as np
```

```python
client = pymongo.MongoClient("localhost",connect=False)
db = client["Zufang-beijing-530"]
table = db["zufang"]
data = pd.DataFrame(list(table.find()))
del data["_id"]
del data["更新时间"]
df = data.drop_duplicates()
df.tail(3)
```

```python
def trans_int(x):
    try:
        return int(x)
    except:
        return np.nan
df = df.iloc[:,[1,2,3,6,7,8]].dropna()
df["建筑面积"] =df["建筑面积"].apply(trans_int)
df.head()
```

```python
df = df.rename(columns={"地址":"name","建筑面积":"area","户型":"house_type","租金":"price","网址":"url","楼层":"floor"})
df.tail(3)
```

```python
df["average_price"] = (df["price"]/df["area"]).apply(trans_int)
df.tail(3)
```

```python
df = df[df.average_price>50]
df = df[df.average_price<70]
df.sort_values("area")
```

```python
df = df[df.area<80].reset_index()
df.tail(3)
```

```python
del df["index"]
```

```python
df.to_csv("zufang_me.csv",encoding="utf-8")
```
